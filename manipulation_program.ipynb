Importing libraries

```python
import pandas as pd
import re
import PySimpleGUI as sg
import random
```

Getting data

```python
data = pd.read_csv("recipes.csv")
data.head(10)
```

Dealing with missing values

```python
data.drop(data[data.ingredients == '[]'].index, inplace=True)
len(data)
```

Some data manipulation

```python
data['name_ingredients'] = data.ingredients.apply(lambda x: [i.split(':')[0].strip() for i in re.sub("[\[\]'\"]", "", x).split(',')])
data['type_dish'] = data.type_dish.str.strip()
data['recipe_steps'] = data.recipe_steps.apply(lambda x: [re.sub("'", "", i).strip() for i in re.sub("[\[\]\"]", "", x).split("',")])
data
```

```python
data.to_csv('recipes.csv', index=False)
```

Getting list of ingridients

```python
list_ingredient = []
for i in data.name_ingredients:
    list_ingredient += i
list_ingredient = set(list_ingredient)
```

Program

```python
chosen_ingredients = []

sg.theme('LightPurple') 

layout = [[sg.Text('Choose the type'),
           sg.Combo(list(data.type_dish.unique()), size=(30, 8), key="t", default_value="Сніданок")],
          [sg.Text('Choose the ingridients'),
           sg.Input(size=(27, 2), enable_events=True, key='-INPUT-')],
          [sg.Listbox(list_ingredient, size=(35, 8), enable_events=True, select_mode="multiple", key="-LIST-")],
          [sg.Text('Choosen ingridients')],
          [sg.Listbox(chosen_ingredients, size=(35, 8), enable_events=True, select_mode="multiple", key="-L-")],
          [sg.Button('Show'), sg.Button('Reset'), sg.Button('Exit')]]

window = sg.Window('Random recipe', layout, size=(450,450))
# Event Loop
while True:
    event, values = window.read()
    if event in (sg.WIN_CLOSED, 'Exit'):                # always check for closed window
        break
    if values['-INPUT-'] != '':                         # if a keystroke entered in search field
        search = values['-INPUT-']
        new_values = [x for x in list_ingredient if search in x]  # do the filtering
        window["-LIST-"].update(new_values)     # display in the listbox
    else:
        # display original unfiltered list
        window["-LIST-"].update(list_ingredient)
    # if a list item is chosen
    if event == '-LIST-' and len(values['-LIST-']):
        chosen_ingredients += values['-LIST-']
        window['-L-'].update(chosen_ingredients)
    if event == "Show":
        data_current = data
        for ingr in chosen_ingredients:
            data_current = data_current[data_current.ingredients.str.contains(ingr)]
        recipe_row = data_current.iloc[random.randint(0, len(data_current)-1)]
        ingred = [i.strip() for i in re.sub("[\[\]'\"]", "", recipe_row['ingredients']).split(',')]
        sg.Popup(recipe_row["name"] + "\n\nIngredients:\n" + '\n'.join(ingred) + "\n\nSteps:\n" + re.sub("\\\\n", " ", '\n'.join(recipe_row["recipe_steps"])))
    if event == "Reset":
        window["t"].update("")
        window["-LIST-"].update(False)
        window["-L-"].update(True)
        chosen_ingredients = []
        
window.close()
```
